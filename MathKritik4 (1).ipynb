```python
import matplotlib.pyplot as plt
import numpy as np

def gradient_descent(f, learning_rate, initial_point):
    def deriv(f, base_point):  # Estimate the derivative of the function f at base_point using symmetric approximation
        return (f(base_point + 10**-10) - f(base_point - 10**-10)) / (2 * 10**-10)

    x_coords = [initial_point]  # This list stores the x_n values
    y_coords = [f(initial_point)]  # This list stores the y_n values

    i = 0
    while np.abs(deriv(f, x_coords[i])) > 1e-10:  # Convergence condition
        xi = x_coords[i] - learning_rate * deriv(f, x_coords[i])  # Gradient descent step
        x_coords.append(xi)
        y_coords.append(f(xi))
        i += 1

    # Plotting portion
    plot_range = np.linspace(
        min(x_coords) - 0.5,  # Adjust plot range for visualization
        max(x_coords) + 0.5, 
        10000
    )
    function_range = [f(i) for i in plot_range]
    plt.plot(plot_range, function_range, label="f(x)")  # Plot f(x)
    plt.plot(x_coords, y_coords, marker="o", label="GD Path")  # Plot GD points
    plt.legend()
    plt.title("Gradient Descent")
    plt.xlabel("x")
    plt.ylabel("f(x)")
    plt.grid()
    plt.show()

    return round(x_coords[-1], 3), round(y_coords[-1], 3)  # Return the last x_n and y_n

# Function definitions
def first_function(x):
    return x**2

def second_function(x):
    return x**4 - 2 * x**2

def third_function(x):
    if x > 0:
        return x**x
    elif x == 0:
        return 1
    else:
        return abs(x)**abs(x)

def fourth_function(x):
    return abs(x)

# Run gradient descent for each function
print(gradient_descent(first_function, 0.3, 2))
print(gradient_descent(second_function, 0.3, -1))
print(gradient_descent(third_function, 0.3, 1))
print(gradient_descent(fourth_function, 0.3, 2))
```
